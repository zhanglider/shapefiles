```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import geopandas as gpd
%pylab inline
```

```python
taxlots = gpd.read_file('DTM_Tax_Lot_Polygon.shp')
```

```python
taxlots
```

```python
#clean the data
taxlots['BORO'] = pd.to_numeric(taxlots['BORO'])
taxlots.dtypes
```

```python
fig = plt.figure(figsize=[10,13])
ax = fig.add_subplot(111)
ax.set_aspect('equal')
taxlots.plot(ax=ax)
```

```python
resi = pd.read_csv('resi.csv')
resi
```

```python
taxlots_resi = pd.merge(taxlots, resi, left_on=['BORO', 'BLOCK', 'LOT'], right_on=['Boro', 'Block', 'Lot'])
taxlots_resi = taxlots_resi.sort_values(by='% over GHGI threshold', ascending = False)
taxlots_resi
```

```python
fig = plt.figure(figsize=[100,130])
ax = fig.add_subplot(234)
ax.set_aspect('equal')
taxlots.plot(ax=ax, color='grey')
taxlots_resi.plot(ax=ax, column='Retrofit depth', legend = True)
```

```python
fig.savefig('retrofit.png')
```
