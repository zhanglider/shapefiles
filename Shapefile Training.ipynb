```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import geopandas as gpd
%pylab inline
```

```python
manhattan = gpd.read_file('MNMapPLUTO.shp')
mih = gpd.read_file('nycmih_20181030.shp')
```

```python
manhattan
```

```python
fig = plt.figure(figsize=[10,13])
ax = fig.add_subplot(111)
ax.set_aspect('equal')
manhattan.plot(ax=ax)
```

```python
mih.head()
```

```python
fig = plt.figure(figsize=[10,13])
ax = fig.add_subplot(111)
ax.set_aspect('equal')
mih.plot(ax=ax)
```

```python
mih = mih.to_crs(manhattan.crs)
mih.plot()
```

```python
fig = plt.figure(figsize=[10,13])
ax = fig.add_subplot(111)
ax.set_aspect('equal')
manhattan.plot(cmap='Blues_r',ax=ax,alpha=1)
mih.plot(ax=ax, color='yellow')
```
